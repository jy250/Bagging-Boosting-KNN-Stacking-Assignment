```python
#1. What is the fundamental idea behind ensemble techniques? How does bagging differ from boosting in terms of approach and objective
"""The fundamental idea behind ensemble techniques is to combine multiple models to improve overall prediction accuracy and reduce errors compared to a single model. By aggregating the outputs of several weak or base learners, ensembles achieve better generalization and stability. Bagging (Bootstrap Aggregating) works by training multiple models independently on different random subsets of the training data (with replacement) and then averaging or voting their predictions to reduce variance. In contrast, Boosting builds models sequentially, where each new model focuses on correcting the errors made by the previous ones, thereby reducing bias. Bagging aims to decrease variance, while Boosting aims to decrease bias and create a strong learner from weak ones.
"""
```

```python
#2. Explain how the Random Forest Classifier reduces overfitting compared to a single decision tree. Mention the role of two key hyperparameters in this process
"""A Random Forest Classifier reduces overfitting by building multiple decision trees on different random subsets of the data and averaging their predictions, which smooths out individual tree errors. Unlike a single decision tree that can memorize training data, the randomness in data sampling and feature selection makes Random Forest more generalizable. Two key hyperparameters play a major role:

n_estimators – the number of trees in the forest; more trees reduce variance and improve stability.

max_features – the number of features considered for splitting at each node; limiting this introduces randomness and ensures trees are less correlated, further reducing overfitting.
"""
```

```python
#3.What is Stacking in ensemble learning? How does it differ from traditional bagging/boosting methods? Provide a simple example use case
"""Stacking (Stacked Generalization) is an ensemble learning technique where multiple different models (called base learners) are trained, and their predictions are combined using a meta-model (or blender) that learns the best way to combine them. Unlike bagging (which averages predictions of similar models trained on random samples) or boosting (which builds models sequentially to correct previous errors), stacking combines diverse models (e.g., decision trees, logistic regression, and SVMs) in parallel and uses another model to learn how to best integrate their outputs.

Example use case: In a loan approval prediction system, a stacking ensemble could combine predictions from a decision tree, logistic regression, and random forest, and then use a gradient boosting model as the meta-learner to produce the final decision, improving accuracy over any single model.
"""
```

```python
#4 . What is the OOB Score in Random Forest, and why is it useful? How does it help in model evaluation without a separate validation set
"""The OOB (Out-of-Bag) Score in a Random Forest is an internal validation metric that measures the model’s performance using the data samples not included in the bootstrap sample for each tree. When a tree is trained, roughly one-third of the training data is left out (OOB samples), and these are later used to test that tree’s predictions. This provides an unbiased estimate of the model’s accuracy without needing a separate validation set.

It’s useful because it saves data and computation time—allowing performance evaluation directly during training. The OOB score helps detect overfitting and gives a reliable indication of how well the Random Forest will generalize to unseen data.
"""
```

```python
#5. Compare AdaBoost and Gradient Boosting in terms of:
# How they handle errors from weak learners
# Weight adjustment mechanism
# Typical use cases
"""Here’s a clear comparison between **AdaBoost** and **Gradient Boosting** based on the given points:

| Aspect                                        | **AdaBoost**                                                                                                                        | **Gradient Boosting**                                                                                                                       |
| --------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------- |
| **How they handle errors from weak learners** | Focuses more on misclassified samples by increasing their weights so that the next learner pays more attention to them.             | Learns the residual (error) of the previous model by fitting a new model to minimize the overall loss function.                             |
| **Weight adjustment mechanism**               | Adjusts the **weights of data points** — increases weights for misclassified instances and decreases for correctly classified ones. | Adjusts **model parameters** by computing gradients of the loss function and adding new learners to correct previous errors.                |
| **Typical use cases**                         | Works well for simple classification problems like spam detection or face recognition.                                              | Used for both classification and regression tasks, especially where complex relationships exist (e.g., risk prediction, sales forecasting). |

**In short:** AdaBoost focuses on **data weighting**, while Gradient Boosting focuses on **error correction via gradient descent**.

"""
```

```python
#6. Why does CatBoost perform well on categorical features without requiring extensive preprocessing? Briefly explain its handling of categorical variables
"""CatBoost performs well on categorical features because it uses a built-in, efficient method to handle them without needing manual preprocessing like one-hot encoding or label encoding. Instead, it converts categorical values into numerical representations using target-based statistics (mean encoding) combined with a special technique called ordered boosting.

In this process, CatBoost replaces a categorical value with an average target value calculated from other examples, while carefully preventing target leakage by using only past data during computation. This approach captures useful information from categorical features while maintaining model integrity. As a result, CatBoost handles categorical data naturally and efficiently, reducing preprocessing time and improving accuracy.
"""
```

```python
#7.  -------------------------------------------
# Step 1: Import required libraries
# -------------------------------------------
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# -------------------------------------------
# Step 2: Load the Wine dataset
# -------------------------------------------
wine = load_wine()
X = wine.data
y = wine.target

print("Feature shape:", X.shape)
print("Target classes:", np.unique(y))

# -------------------------------------------
# Step 3: Split data (70% train, 30% test)
# -------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# -------------------------------------------
# Step 4: Train KNN without scaling (default K=5)
# -------------------------------------------
knn_default = KNeighborsClassifier(n_neighbors=5)
knn_default.fit(X_train, y_train)

y_pred_default = knn_default.predict(X_test)

print("\n=== KNN WITHOUT SCALING ===")
print("Accuracy:", accuracy_score(y_test, y_pred_default))
print("Classification Report:\n", classification_report(y_test, y_pred_default))

# -------------------------------------------
# Step 5: Apply StandardScaler and retrain KNN
# -------------------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)

print("\n=== KNN WITH STANDARD SCALING ===")
print("Accuracy:", accuracy_score(y_test, y_pred_scaled))
print("Classification Report:\n", classification_report(y_test, y_pred_scaled))

# -------------------------------------------
# Step 6: Optimize KNN using GridSearchCV
# -------------------------------------------
param_grid = {
    'n_neighbors': list(range(1, 21)),
    'metric': ['euclidean', 'manhattan']
}

grid = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid.fit(X_train_scaled, y_train)

print("\n=== GRID SEARCH RESULTS ===")
print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation Accuracy:", grid.best_score_)

# -------------------------------------------
# Step 7: Evaluate optimized KNN
# -------------------------------------------
best_knn = grid.best_estimator_
y_pred_best = best_knn.predict(X_test_scaled)

print("\n=== OPTIMIZED KNN RESULTS ===")
print("Accuracy:", accuracy_score(y_test, y_pred_best))
print("Classification Report:\n", classification_report(y_test, y_pred_best))

# -------------------------------------------
# Step 8: Summary comparison
# -------------------------------------------
summary = pd.DataFrame({
    "Model": ["Unscaled KNN (k=5)", "Scaled KNN (k=5)", "Optimized KNN"],
    "Accuracy": [
        accuracy_score(y_test, y_pred_default),
        accuracy_score(y_test, y_pred_scaled),
        accuracy_score(y_test, y_pred_best)
    ]
})

print("\n=== PERFORMANCE COMPARISON ===")
print(summary)
```

```python
# 8 .---------------------------------------------------------
# Step 1: Import required libraries
# ---------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# ---------------------------------------------------------
# Step 2: Load the Breast Cancer dataset
# ---------------------------------------------------------
data = load_breast_cancer()
X = data.data
y = data.target
print("Dataset shape:", X.shape)
print("Target classes:", data.target_names)

# ---------------------------------------------------------
# Step 3: Standardize the data
# ---------------------------------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------------------------------------------
# Step 4: Apply PCA
# ---------------------------------------------------------
pca = PCA()
pca.fit(X_scaled)
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Scree Plot (Explained Variance Ratio)
plt.figure(figsize=(8,5))
plt.plot(np.arange(1, len(explained_variance)+1), explained_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot: PCA Explained Variance')
plt.grid(True)
plt.show()

# ---------------------------------------------------------
# Step 5: Retain 95% variance and transform the dataset
# ---------------------------------------------------------
pca_95 = PCA(n_components=0.95)
X_pca = pca_95.fit_transform(X_scaled)
print(f"\nNumber of components retained for 95% variance: {pca_95.n_components_}")

# ---------------------------------------------------------
# Step 6: Train-Test Split
# ---------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
X_train_pca, X_test_pca, _, _ = train_test_split(X_pca, y, test_size=0.3, random_state=42, stratify=y)

# ---------------------------------------------------------
# Step 7: Train KNN on original and PCA-transformed data
# ---------------------------------------------------------
knn_original = KNeighborsClassifier(n_neighbors=5)
knn_original.fit(X_train, y_train)
y_pred_orig = knn_original.predict(X_test)

knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)

# ---------------------------------------------------------
# Step 8: Compare Accuracy
# ---------------------------------------------------------
acc_original = accuracy_score(y_test, y_pred_orig)
acc_pca = accuracy_score(y_test, y_pred_pca)

print("\n=== MODEL PERFORMANCE ===")
print(f"Accuracy (Original Data): {acc_original:.4f}")
print(f"Accuracy (PCA 95% Data): {acc_pca:.4f}")

# ---------------------------------------------------------
# Step 9: Classification Reports
# ---------------------------------------------------------
print("\n--- Original Data ---")
print(classification_report(y_test, y_pred_orig, target_names=data.target_names))

print("\n--- PCA Transformed Data ---")
print(classification_report(y_test, y_pred_pca, target_names=data.target_names))

# ---------------------------------------------------------
# Step 10: Visualize the first two principal components
# ---------------------------------------------------------
pca_2d = PCA(n_components=2)
X_2d = pca_2d.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_2d[:,0], X_2d[:,1], c=y, cmap='coolwarm', edgecolor='k', s=50)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA: First Two Principal Components")
plt.colorbar(label='Class (0=Malignant, 1=Benign)')
plt.show()
```

```python
#9 . ---------------------------------------------------------
# Step 1: Import required libraries
# ---------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# ---------------------------------------------------------
# Step 2: Generate synthetic regression dataset
# ---------------------------------------------------------
X, y = make_regression(
    n_samples=500,
    n_features=10,
    noise=20,
    random_state=42
)

# Split dataset into train (70%) and test (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Scale features (important for distance-based models)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ---------------------------------------------------------
# Step 3: Train KNN Regressor with different distance metrics
# ---------------------------------------------------------

# a. Euclidean distance
knn_euclidean = KNeighborsRegressor(n_neighbors=5, metric='euclidean')
knn_euclidean.fit(X_train_scaled, y_train)
y_pred_euclidean = knn_euclidean.predict(X_test_scaled)
mse_euclidean = mean_squared_error(y_test, y_pred_euclidean)

# b. Manhattan distance
knn_manhattan = KNeighborsRegressor(n_neighbors=5, metric='manhattan')
knn_manhattan.fit(X_train_scaled, y_train)
y_pred_manhattan = knn_manhattan.predict(X_test_scaled)
mse_manhattan = mean_squared_error(y_test, y_pred_manhattan)

print("=== KNN REGRESSOR COMPARISON (K=5) ===")
print(f"Euclidean Distance MSE: {mse_euclidean:.4f}")
print(f"Manhattan Distance MSE: {mse_manhattan:.4f}")

# ---------------------------------------------------------
# Step 4: Analyze Bias-Variance tradeoff (varying K)
# ---------------------------------------------------------
k_values = [1, 5, 10, 20, 50]
mse_values = []

for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k, metric='euclidean')
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# ---------------------------------------------------------
# Step 5: Plot K vs MSE
# ---------------------------------------------------------
plt.figure(figsize=(8,5))
plt.plot(k_values, mse_values, marker='o', linestyle='-', color='b')
plt.title('KNN Regressor: K vs MSE (Bias-Variance Tradeoff)')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()

# ---------------------------------------------------------
# Step 6: Summary
# ---------------------------------------------------------
print("\n=== K vs MSE Summary ===")
for k, mse in zip(k_values, mse_values):
    print(f"K = {k:2d} --> MSE = {mse:.4f}")
```

```python
#10 .Step-by-Step Implementation
# ------------------------------------------------------------
# Step 1: Import libraries
# ------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.decomposition import PCA

# ------------------------------------------------------------
# Step 2: Load the Pima Indians Diabetes dataset
# ------------------------------------------------------------
# You can download from: https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
cols = [
    "Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
    "Insulin", "BMI", "DiabetesPedigree", "Age", "Outcome"
]
df = pd.read_csv(url, names=cols)

print("Dataset shape:", df.shape)
df.head()

#Identify and Replace Invalid Zero Values
# Columns where 0 is invalid and should be treated as missing
invalid_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

df[invalid_cols] = df[invalid_cols].replace(0, np.nan)
print("Missing values before imputation:\n", df.isna().sum())

#KNN Imputation
imputer = KNNImputer(n_neighbors=5)
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=cols)

print("\nMissing values after imputation:\n", df_imputed.isna().sum())

#Split & Standardize Data
X = df_imputed.drop("Outcome", axis=1)
y = df_imputed["Outcome"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train KNN with Different Algorithms
methods = ['brute', 'kd_tree', 'ball_tree']
results = []

for method in methods:
    start = time.time()

    knn = KNeighborsClassifier(n_neighbors=5, algorithm=method)
    knn.fit(X_train_scaled, y_train)

    train_time = time.time() - start
    y_pred = knn.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)

    results.append((method, acc, train_time))

# Display results
df_results = pd.DataFrame(results, columns=['Algorithm', 'Accuracy', 'Training Time (s)'])
print("\n=== Algorithm Performance Comparison ===")
print(df_results)

# Visualization – Decision Boundary (Top 2 Important Features)
# Select two most important features
X_vis = df_imputed[['Glucose', 'BMI']]
y_vis = df_imputed['Outcome']

# Train-test split again for visualization
Xv_train, Xv_test, yv_train, yv_test = train_test_split(
    X_vis, y_vis, test_size=0.3, random_state=42, stratify=y_vis
)

scaler_vis = StandardScaler()
Xv_train_scaled = scaler_vis.fit_transform(Xv_train)
Xv_test_scaled = scaler_vis.transform(Xv_test)

# Choose best-performing algorithm
best_algo = df_results.iloc[df_results['Accuracy'].idxmax()]['Algorithm']
knn_best = KNeighborsClassifier(n_neighbors=5, algorithm=best_algo)
knn_best.fit(Xv_train_scaled, yv_train)

# Create meshgrid for boundary plot
x_min, x_max = Xv_train_scaled[:, 0].min() - 1, Xv_train_scaled[:, 0].max() + 1
y_min, y_max = Xv_train_scaled[:, 1].min() - 1, Xv_train_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = knn_best.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(Xv_train_scaled[:, 0], Xv_train_scaled[:, 1],
            c=yv_train, cmap='coolwarm', edgecolor='k', s=50)
plt.title(f"KNN Decision Boundary ({best_algo.upper()} method)")
plt.xlabel("Glucose (scaled)")
plt.ylabel("BMI (scaled)")
plt.show()
```
